```python
# Purpose: A text-based adventure game to practice functions, conditionals, and loops.
# Developed with assistance from GitHub Copilot.

import sys
```

```python
def start_game():
    """Introduces the quest and handles the first major choice."""
    print("Welcome to the Ancient Land Quest!")
    player_name = input("Enter your name, brave explorer: ")
    print(f"\nHello {player_name}! You stand at a crossroad.")

    choice = input("Do you want to explore the 'forest' or enter the 'cave'? ").lower()

    if choice == 'forest':
        forest_path()
    elif choice == 'cave':
        cave_path()
    else:
        print("Invalid choice. The ground shakes and you must choose again.")
        start_game()
```

```python
def forest_path():
    """Handles the forest scenario and decisions."""
    print("\nYou enter a dense, dark forest. Birds stop singing as you pass.")
    choice = input("Do you want to 'follow' the river or 'climb' a tree? ").lower()

    if choice == 'follow':
        print("\nYou follow the river and find a hidden chest filled with gold!")
        print("CONGRATULATIONS! You found the treasure.")
        end_game()
    elif choice == 'climb':
        print("\nYou climb the tree but a branch breaks! You fall and must retreat.")
        print("GAME OVER: Your adventure ends here.")
        end_game()
    else:
        print("Confused, you wander in circles.")
        forest_path()
```

```python
def cave_path():
    """Handles the cave scenario and decisions."""
    print("\nThe cave is cold and smells of damp earth.")
    choice = input("Do you want to 'light' a torch or 'proceed' in the dark? ").lower()

    if choice == 'light':
        print("\nThe light reveals a sleeping dragon. You sneak past and find the treasure!")
        print("WINNER: You are legendary!")
        end_game()
    elif choice == 'proceed':
        print("\nIn the dark, you trip into a deep pit.")
        print("GAME OVER: Better luck next time.")
        end_game()
    else:
        print("The darkness is overwhelming. Try again.")
        cave_path()
```

```python
def end_game():
    """Handles the restart logic."""
    choice = input("\nWould you like to play again? (yes/no): ").lower()
    if choice == 'yes':
        start_game()
    else:
        print("Thank you for playing! Goodbye.")
        sys.exit()
```

```python

if __name__ == "__main__":
    # Task 5: Call start_game to begin the adventure
    start_game()
```
